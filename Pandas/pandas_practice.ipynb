```python
# Dependencies and Setup
import pandas as pd
import statistics as stat
```

```python
# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
school_data.tail()
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#Create binary statistic for passing math and reading based on if a students score is higher than 59
passingr = []
passingm = []
for student in school_data_complete['math_score']:
    if student >= 60:
        passingm.append(1)
    else:
        passingm.append(0)
school_data_complete['passing_math'] = passingm

passingr = []
for student in school_data_complete['reading_score']:
    if student >= 60:
        passingr.append(1)
    else:
        passingr.append(0)
school_data_complete['passing_reading'] = passingr
school_data_complete.head()
```

```python
#District summary 
#Total Schools
total_schools = len(school_data['school_name'])
#Total Students
total_students = sum(school_data['size'])
#Total Budget
total_budget = sum(school_data['budget'])
#Average Math Score
avg_math = sum(school_data_complete['math_score'])/total_students
#Average Reading Score
avg_reading= sum(school_data_complete['reading_score'])/total_students
#% Passing Math
pass_rate_math = round(sum(school_data_complete['passing_math'])/total_students*100,2)
#% Passing Reading
pass_rate_reading = round(sum(school_data_complete['passing_reading'])/total_students*100,2)
#Overall Passing Rate (Average of the above two)
district_overall_pass_rate = (pass_rate_reading+pass_rate_math)/2

#Create table of each statistic
district_summary = pd.DataFrame([district_overall_pass_rate])
district_summary['Total Schools'] = [total_schools]
district_summary['Total Students'] = [total_students]
district_summary['Total Budget'] = [total_budget]
district_summary['Average Math Score'] = [avg_math]
district_summary['Average Reading Score'] = [avg_reading]
district_summary['Math Pass Rate'] = [pass_rate_math]
district_summary['Reading Pass Rate'] = [pass_rate_reading]
district_summary['Overall Pass Rate'] = [district_overall_pass_rate]
district_summary = district_summary.drop(columns=[0])
district_summary

```

```python
#School summary
#group the data by school name
gschools = school_data_complete.groupby(['school_name'])

#School Type
school_type = gschools['type'].first()
#Total Students
school_size = gschools['size'].count()
#Total School Budget
school_budget = gschools['budget'].sum()/school_size
#Per Student Budget
per_student = school_budget/school_size
#Average Math Score
math_school = gschools['math_score'].sum()
avg_math_school = math_school/school_size
#Average Reading Score
reading_school = gschools['reading_score'].sum()
avg_reading_school = reading_school/school_size
#% Passing Math
school_passm = gschools['passing_math'].sum()
school_math_rate = school_passm/school_size*100

#% Passing Reading
school_passr = gschools['passing_reading'].sum()
school_reading_rate = school_passr/school_size*100
#Overall Passing Rate (Average of the above two)
school_overall_pass_rate = (school_math_rate+school_reading_rate)/2

#Create a table of each statistic
school_summary = pd.DataFrame(school_overall_pass_rate)
school_summary['Size'] = round(school_size, 2)
school_summary['Type'] = school_type
school_summary['Budget'] = school_budget
school_summary['Per Student Budget'] = per_student
school_summary['Average Math Score'] = round(avg_math_school,2)
school_summary['Average Reading Score'] = round(avg_reading_school,2)
school_summary['Math Pass Rate'] = round(school_math_rate,2)
school_summary['Reading Pass Rate'] = round(school_reading_rate,2)
school_summary['Overall Pass Rate'] = round(school_overall_pass_rate,2)
school_summary = school_summary.drop(columns=[0])
school_summary
```

```python
#Top 5 schools
sorted_schools_top = school_summary.sort_values(by = ['Overall Pass Rate'], ascending = False)
top_5 = sorted_schools_top[0:5]
top_5
 
```

```python
#bottom 5 schools
sorted_schools_bottom = school_summary.sort_values(by = ['Overall Pass Rate'], ascending = True)
bottom_5 = sorted_schools_bottom[0:5]
bottom_5
```

```python
#Math scores by grade
#group data by grade
ggrade = school_data_complete.groupby(['school_name', 'grade'])
#Size of each grade
grade_size = ggrade['size'].count()
#Math scores for each grade
math_grade = ggrade['math_score'].sum()
avg_math_grade = math_grade/grade_size

#Create table 
grade_summary_math = pd.DataFrame(avg_math_grade)
grade_summary_math['Average Math Score'] = round(avg_math_grade,2)
grade_summary_math = grade_summary_math.drop(columns=[0])
grade_summary_math
```

```python
#Reading scores by grade
#Average reading scores by grade
reading_grade = ggrade['reading_score'].sum()
avg_reading_grade = reading_grade/grade_size

#Create a table of reading scores
grade_summary_reading = pd.DataFrame(avg_reading_grade)
grade_summary_reading['Average Reading Score'] = round(avg_reading_grade,2)
grade_summary_reading = grade_summary_reading.drop(columns=[0])
grade_summary_reading
```

```python
#Scores by school spending
#Cerate 4 bins for seperate budget levels
budgets = pd.cut(school_data_complete['budget']/school_data_complete['size'],4, labels = ['High Budget' , 
                        'Medium-High Budget',
                         'Medium-Low Budget', 'Low Budget'])
#seperate budget dataframe
budget_df = school_data_complete
#add budget levels to the dataframe
budget_df['Budget Level'] = budgets
#group based on budget level
grouped_budget = budget_df.groupby(['Budget Level'])
#size of each budget level
budget_size = grouped_budget['size'].count()
#math scores for each budget level
budget_math = grouped_budget['math_score'].sum()/budget_size
#reading scores for each budget level
budget_reading = grouped_budget['reading_score'].sum()/budget_size
#math pass rate for each budget level
budget_passm = grouped_budget['passing_math'].sum()
budget_math_rate = budget_passm/budget_size*100
#reading pass rate for each budget level
budget_passr = grouped_budget['passing_reading'].sum()
budget_reading_rate = budget_passr/budget_size*100
#overall pass rate for each budget level
budget_overall_pass_rate = (budget_math_rate+budget_reading_rate)/2

#Create a table of each statistic
budget_summary = pd.DataFrame(budget_overall_pass_rate)
budget_summary['Average Math Score'] = round(budget_math,2)
budget_summary['Average Reading Score'] = round(budget_reading,2)
budget_summary['Math Pass Rate'] = round(budget_math_rate,2)
budget_summary['Reading Pass Rate'] = round(budget_reading_rate,2)
budget_summary['Overall Pass Rate'] = round(budget_overall_pass_rate,2)
budget_summary = budget_summary.drop(columns=[0])
budget_summary

```

```python
#Scores by school size
#Create bins for schools of different sizes
sizes = pd.cut(school_data_complete['size'],3, labels = ['Large' , 
                        'Medium',
                         'Small'])
#seperate dataframe based of the original
size_df = school_data_complete
#add size group statisitic to the dataframe
size_df['Size Level'] = sizes
#group based on the size level
grouped_size = budget_df.groupby(['Size Level'])
#size of each size level
size_size = grouped_size['size'].count()
#math scores for each size level
size_math = grouped_size['math_score'].sum()/size_size
#reading scores for each size level
size_reading = grouped_size['reading_score'].sum()/size_size
#math pass rate for each size level
size_passm = grouped_size['passing_math'].sum()
size_math_rate = size_passm/size_size*100
#reading pass rate for each size level
size_passr = grouped_size['passing_reading'].sum()
size_reading_rate = size_passr/size_size*100
#overall pass rate for each size level
size_overall_pass_rate = (size_math_rate+size_reading_rate)/2
#Create a table of each statistic
size_summary = pd.DataFrame(size_overall_pass_rate)
size_summary['Average Math Score'] = round(size_math,2)
size_summary['Average Reading Score'] = round(size_reading,2)
size_summary['Math Pass Rate'] = round(size_math_rate,2)
size_summary['Reading Pass Rate'] = round(size_reading_rate,2)
size_summary['Overall Pass Rate'] = round(size_overall_pass_rate,2)
size_summary = size_summary.drop(columns=[0])
size_summary
```

```python
#Scores by school type
#Seperate datframe for school types
type_df = school_data_complete
#group by school type
grouped_type = budget_df.groupby(['type'])
#size of each school type
type_size = grouped_type['size'].count()
#math scores for each school type
type_math = grouped_type['math_score'].sum()/type_size
#reading scores for each school type
type_reading = grouped_type['reading_score'].sum()/type_size
#math pass rate for each school type
type_passm = grouped_type['passing_math'].sum()
type_math_rate = type_passm/type_size*100
#reading pass rate for each school type
type_passr = grouped_type['passing_reading'].sum()
type_reading_rate = type_passr/type_size*100
#overall pass rate for each school type
type_overall_pass_rate = (type_math_rate+type_reading_rate)/2
#create a table of each statistic
type_summary = pd.DataFrame(type_overall_pass_rate)
type_summary['Average Math Score'] = round(type_math,2)
type_summary['Average Reading Score'] = round(type_reading,2)
type_summary['Math Pass Rate'] = round(type_math_rate,2)
type_summary['Reading Pass Rate'] = round(type_reading_rate,2)
type_summary['Overall Pass Rate'] = round(type_overall_pass_rate,2)
type_summary = type_summary.drop(columns=[0])
type_summary
```
